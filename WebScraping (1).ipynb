```python
from bs4 import BeautifulSoup
import requests
```

```python
url='https://en.wikipedia.org/wiki/List_of_international_cricket_centuries_by_Virat_Kohli'
page=requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
soup.find_all('table')[1]
```

```python
# Find all tables with the class 'wikitable sortable plainrowheaders'
tables = soup.find_all('table', class_='wikitable sortable plainrowheaders')

# Select the second table (index 1)
second_table = tables[0]

# Extract header titles from the second table
table2_titles = [th.get_text(strip=True) for th in second_table.find_all('th')]

test_titles=table2_titles[:11]
test_titles
```

```python
import pandas as pd
```

```python
df =pd.DataFrame(columns=test_titles)
df
```

```python
col_data=second_table.find_all('tr')
```

```python
for row in col_data[1:]:
    indi_row_data = []
    runs_data = row.find('th', scope='row')    
    if runs_data:
        indi_row_data.append(runs_data.get_text(strip=True))
    row_data = row.find_all('td')
    indi_row_data.extend([data.get_text(strip=True) for data in row_data])
    length=len(df)
    df.loc[length]=indi_row_data
```
